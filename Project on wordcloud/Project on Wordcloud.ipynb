```python
import re
import string
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
```

```python
with open("C:\\Users\\hp\\Desktop\\Additional Projects\\Imarticus Learning.txt","r") as f:
    test=f.readlines()
```

```python
test
```

```python
txt=[i.strip()for i in test
    ]
print(txt)
```

```python
txt1=' '.join(txt)
print(txt1)
```

```python
txt2=txt1.lower()
print(txt2)
```

```python
txt3=re.sub(r'[^\w\s]','',txt2)
print(txt3)
```

```python
txt4=re.sub('r\d+',' ',txt3)
print(txt4)
```

```python
token=word_tokenize(txt4)
token
```

```python
from nltk.corpus import stopwords
```

```python
txt5=[i for i in token if i not in stopwords.words('english')]
print(txt5)
```

```python

txt5=' '.join(str(x) for x in txt5)
print(txt5)
```

```python
txt6=' '.join(i for i in txt5.split() if not (i.isalpha() and len(i)==1))
txt6
```

```python
txt6=txt6.strip()
txt6
```

```python
words_dict={}
for i in txt6.split():
    words_dict[i]=words_dict.get(i,0)+1
for j in sorted(words_dict):
    print('{}:{}'.format(j,words_dict[j]))
```

```python

word_cloud=WordCloud().generate(txt6)
```

```python
import matplotlib.pyplot as plt
plt.imshow(word_cloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```

```python
wordcloud=WordCloud(width=300,height=300,colormap='Purples_r').generate(txt6)
plt.figure()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
